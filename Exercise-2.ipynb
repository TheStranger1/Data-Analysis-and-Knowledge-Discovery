```python
#1. Load the Iris dataset from http://archive.ics.uci.edu/ml/datasets/Iris

import pandas as pd

# Loading Dataset

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv(url, names=names)
iris.head()
```

```python
# Calculate the mean, median and standard deviation for all attributes of Iris setosa.

iris_setosa=iris.iloc[0:50]

# Mean

iris_setosa.mean()

# median

iris_setosa.median()

# standard deviation

iris_setosa.std()
```

```python
# 3. Select one attribute for species Iris setosa. Plot four histograms of the attribute using the methods
# intoduced in the lectures (Sturges, Scott, Square root, and Freedman-Diaconis) to determine the number of
# bins. Compare the results.

#Next we would like to see if the distributions of the values for different flower species vary in some way.
#Plot histograms of your attribute for Iris virginica and Iris versicolor. Do these histograms give any indication
#about the feasibility or infeasibility of classifying the flower species?

#sturges rule

import matplotlib.pyplot as plt 
iris_setosa['sepal-length'].plot(kind= 'hist', bins=7)
plt.xlabel('sepal length(cm)')
plt.show()

# Scott's rule

iris_setosa['sepal-length'].plot(kind= 'hist', bins=5)
plt.xlabel('sepal length(cm)')
plt.show()

# Square Root 
# Square root=7.07

iris_setosa['sepal-length'].plot(kind= 'hist', bins=7)
plt.xlabel('sepal length(cm)')
plt.show()

# Freedman-Diaconis

iris_setosa['sepal-length'].plot(kind= 'hist', bins=7)
plt.xlabel('sepal length(cm)')
plt.show()

# Iris-versicolor

iris_versicolor = iris.iloc[50:100]
iris_versicolor['sepal-length'].plot(kind= 'hist')
plt.xlabel('sepal length(cm)')
plt.show()


#Iris-verginica

iris_verginica = iris.iloc[100:150]
iris_verginica['sepal-length'].plot(kind= 'hist')
plt.xlabel('sepal length(cm)')
plt.show()

```

```python
# 4. Produce boxplot plots of attributes. Can you see any outliers?

# Sepal-length

iris.plot(y= 'sepal-length', kind= 'box')
plt.ylabel('sepal length(cm)')
plt.show()

# Sepal-width

iris.plot(y= 'sepal-width', kind= 'box')
plt.ylabel('sepal width(cm)')
plt.show()

# Petal-length

iris.plot(y= 'petal-length', kind= 'box')
plt.ylabel('petal length(cm)')
plt.show()

# Petal-width

iris.plot(y= 'petal-width', kind= 'box')
plt.ylabel('petal width(cm)')
plt.show()
```
